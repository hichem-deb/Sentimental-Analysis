```
res = [] 
for val in list: 
    if val != None : 
        res.append(val) 
```

```
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid= SentimentIntensityAnalyzer()
```

```
l1=[]
l2=[]
word=[]
new_list=[]
for l in res:
  print(l)
  word_list = str(l).split()
  print(word_list)
  dict=sid.polarity_scores(str(l))
  print(dict)
  print(max(dict, key=dict.get))

  new_list.append([l,max(dict, key=dict.get)])
  l1.append(l)
  l2.append(max(dict, key=dict.get))

  word.append(word_list)





  
```

```
import numpy as np

percentile_list = pd.DataFrame(
    {'Text': l1,
     'Sentiment': l2
    })
percentile_list = percentile_list[percentile_list.Sentiment != 'compound']
percentile_list
```

```
number_neu = l2.count('neu')
number_neu
```

```
percent_neu= number_neu/len(percentile_list['Sentiment']) * 100
percent_neu
```

```
number_pos = l2.count('pos')
number_pos
```

```
percent_pos= number_pos/len(percentile_list['Sentiment']) * 100
percent_pos
```

```
number_neg = l2.count('neg')
number_neg
```

```
percent_neg= number_neg/len(percentile_list['Sentiment']) * 100
percent_neg
```

```
import matplotlib.pyplot as plt 
  
sentiments = ['Neutral Comments', 'Positive Comments', 'Negative Comments'] 

slices = [percent_neu, percent_pos, percent_neg] 
 
colors = ['y', 'b', 'r'] 
  
plt.pie(slices, labels = sentiments, colors=colors,  
        startangle=90, shadow = True, 
        radius = 3, autopct = '%1.1f%%') 
  

plt.legend() 

plt.show() 
```
